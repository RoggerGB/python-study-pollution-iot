```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


dates = pd.date_range(start="2024-06-06", end="2024-06-10 23:00:00", freq='H')
num_hours = len(dates)


np.random.seed(0)  # Para reproducibilidad
pm25 = np.random.uniform(low=5, high=55, size=num_hours)
pm10 = np.random.uniform(low=50, high=200, size=num_hours)
so2_ugm3 = np.random.uniform(low=5, high=20, size=num_hours)
no2_ugm3 = np.random.uniform(low=20, high=40, size=num_hours)
co_ugm3 = np.random.uniform(low=180, high=220, size=num_hours)


data = {
    'Fecha': dates,
    'PM 2.5 (µg/m³)': pm25,
    'PM 10 (µg/m³)': pm10,
    'SO2 (ppm)': so2_ugm3,
    'NO2 (ppm)': no2_ugm3,
    'CO (ppm)': co_ugm3
}

df = pd.DataFrame(data)
```

# GRAFICO POR PM 2.5

```python

df1_avg = df_final.query("hora > '08:00:00' and hora<'17:00:00'").groupby('hora')['PM 2.5 (µg/m³)'].mean().reset_index()
df2_avg = df_senamhi.query("hora_formato > '08:00:00' and hora_formato<'17:00:00'").groupby('hora_formato')['PM 2.5 µg / m 3'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(df1_avg['hora'], df1_avg['PM 2.5 (µg/m³)'], marker='o', linestyle='-', label='Data propia')
plt.plot(df2_avg['hora_formato'], df2_avg['PM 2.5 µg / m 3'], marker='o', linestyle='-', label='Senamhi')
plt.xlabel('Hora')
plt.ylabel('PM Promedio')
plt.title('Comparación de PM2.5 Promedio por Hora')
plt.legend()
plt.grid(True)
plt.show()
```
