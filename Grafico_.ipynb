```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear índices únicos para cada punto para que las horas repetidas se puedan graficar
df1['Index'] = range(len(df1))
df2['Index'] = range(len(df2))

# Crear el gráfico
plt.figure(figsize=(12,6))

# Ancho de las barras
bar_width = 0.4

# Posición de las barras en el eje x
r1 = range(len(df1))
r2 = [x + bar_width for x in r1]

# Graficar el PM2.5 del primer dataframe como barras
plt.bar(r1, df1['PM2.5'], width=bar_width, label='PM2.5 Dataset 1', color='blue')

# Graficar el PM2.5 del segundo dataframe como barras
plt.bar(r2, df2['PM2.5'], width=bar_width, label='PM2.5 Dataset 2', color='green')

# Añadir etiquetas y título
plt.xlabel('Observación')
plt.ylabel('PM2.5')
plt.title('Comparativo de PM2.5 entre dos Dataframes (horas repetidas)')
plt.xticks([r + bar_width / 2 for r in r1], df1['hora'], rotation=45)

# Modificar el rango del eje y
plt.ylim(30, 150)  # Ajusta el rango del eje y según tus necesidades

# Añadir leyenda
plt.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Asegurarse de que la columna 'hora' sea de tipo string
df1['hora'] = df1['hora'].astype(str)
df2['hora'] = df2['hora'].astype(str)

# Crear índices únicos para cada punto para que las horas repetidas se puedan graficar
df1['Index'] = range(len(df1))
df2['Index'] = range(len(df2))

# Crear el gráfico
plt.figure(figsize=(14,8))

# Definir los segmentos de índices y colores
segments = [(0, 9), (10, 14), (15, len(df1) - 1)]
colors = ['blue', 'green', 'red']  # Colores para los segmentos

# Graficar el PM2.5 del primer dataframe por segmentos
for (start, end), color in zip(segments, colors):
    df_filtrado = df1[(df1['Index'] >= start) & (df1['Index'] <= end)]
    plt.plot(df_filtrado['Index'], df_filtrado['PM2.5'], color=color, label=f'Segmento {start}-{end} Dataset 1', marker='o')

# Graficar el PM2.5 del segundo dataframe por segmentos
for (start, end), color in zip(segments, colors):
    df_filtrado = df2[(df2['Index'] >= start) & (df2['Index'] <= end)]
    plt.plot(df_filtrado['Index'], df_filtrado['PM2.5'], color=color, linestyle='--', label=f'Segmento {start}-{end} Dataset 2', marker='x')

# Añadir etiquetas y título
plt.xlabel('Índice')
plt.ylabel('PM2.5')
plt.title('Comparativo de PM2.5 entre dos Dataframes por Hora')

# Configurar el eje x para mostrar las horas
plt.xticks(df1['Index'], df1['hora'], rotation=45, ha='right')

# Añadir leyenda
plt.legend(loc='upper left', bbox_to_anchor=(1,1))

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear índices únicos para cada punto para que las horas repetidas se puedan graficar
df1['Index'] = range(len(df1))
df2['Index'] = range(len(df2))

# Crear el gráfico
plt.figure(figsize=(12,6))

# Definir los segmentos de índices y colores
segments = [(0, 9), (10, 14), (15, len(df1) - 1)]
colors = ['blue', 'green', 'red']  # Colores para los segmentos

# Concatenar datos para mantener la continuidad
df1_combined = pd.concat([df1[(df1['Index'] >= start) & (df1['Index'] <= end)] for (start, end) in segments])
df2_combined = pd.concat([df2[(df2['Index'] >= start) & (df2['Index'] <= end)] for (start, end) in segments])

# Graficar el PM2.5 del primer dataframe por segmentos
for (start, end), color in zip(segments, colors):
    df_filtrado = df1_combined[(df1_combined['Index'] >= start) & (df1_combined['Index'] <= end)]
    plt.plot(df_filtrado['Index'], df_filtrado['PM2.5'], color=color, marker='o', linestyle='-')

# Graficar el PM2.5 del segundo dataframe por segmentos
for (start, end), color in zip(segments, colors):
    df_filtrado = df2_combined[(df2_combined['Index'] >= start) & (df2_combined['Index'] <= end)]
    plt.plot(df_filtrado['Index'], df_filtrado['PM2.5'], color=color, linestyle='--', marker='x')

# Añadir etiquetas y título
plt.xlabel('Observación')
plt.ylabel('PM2.5')
plt.title('Comparativo de PM2.5 entre dos Dataframes (horas repetidas)')

# Configurar el eje x para mostrar las horas
plt.xticks(df1['Index'], df1['hora'], rotation=45)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
